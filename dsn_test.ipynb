```python
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import MinMaxScaler
from catboost import CatBoostClassifier
import lightgbm as lgb
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier

scaler = MinMaxScaler()
scaled_x = scaler.fit_transform(X)
scaled_t = scaler.transform(T)
skfold = StratifiedKFold(n_splits=30, shuffle=True, random_state=1)
cat = CatBoostClassifier(random_state=1)
rf = RandomForestClassifier(random_state=1)
lgbc = lgb.LGBMClassifier(random_state=1)

# define the voteclass function
```

```python
def voteclass(scaled_t, scaled_x, y):
    scores, preds = [], []
    i = 1
    for train_split, test_split in skfold.split(scaled_x, y):
        x_train, x_test, y_train, y_test = scaled_x[train_split], scaled_x[test_split], y[train_split], y[test_split]
        vc = VotingClassifier(estimators=[('cat_clf',cat), ('random_clf',rf), ('lgb_clf', lgbc)], voting='soft')
        vc.fit(x_train, y_train)
        pred = vc.predict_proba(scaled_t)[:,1]
        preds.append(pred)
        final_predict = np.mean(preds, axis=0)
        return final_predict
```

```python
pred_c = voteclass(scaled_t, scaled_x, y)
```
